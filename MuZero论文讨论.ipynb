Model的3个部分
1. representation function $h$
1. dynamics function $g$
1. prediction function $f$


Monte-Carlo tree search需要的信息

$$
\{N(s,a),Q(s,a),P(s,a),R(s,a),S(s,a)\}
$$

* 其中较为难获得的：policy $P$, reward $R$, and state transition $S$
* Model因此提供：
    * $r^l,s^l=g_\theta(s^{l-1},a^l)$
    * $\mathbf{p}^l,v^l=f_\theta(s^l)$

The MCTS algorithm
outputs a recommended policy $\pi_t$ and estimated value $\nu_t$. An action $a_{t+1}\sim\pi_t$ is then selected.

特点：
* improved policy targets
* improved value targets

**Training**

1. initial step为获取历史（past）observations，执行
$$s^0=h_\theta(o_1,\dotsc,o_t)$$
1. unrolled for $K$ hypothetical steps and aligned to sequences sampled from
the trajectories generated by the MCTS actors
    * MCTS中为：$r^k,s^k=g_\theta(s^{k-1},a^k)$
    * Training中，上式的$a^k$'s替换为real actions $a_{t+1},\dotsc,a_{t+k}$

1. 可以利用的信息：Each observation $o_t$ along the sequence also has a corresponding MCTS policy $\pi_t$, estimated value $\nu_t$ and
environment reward $u_t$.
    * 上述信息中可以直接利用的：$\pi_{t+k}$和$u_{t+k}$


1. unrolled for $K$ hypothetical steps后，model这边为：$\mathbf{p}^{t+k},v^{t+k}=f_\theta(s^{t+k})$
    * $v^t_k$（predicted value）的目标为$z_{t+k}$（value target）
    * $z_{t+k}$为*bootstrapping* n steps into the future from the search value
    * 形式：$z_t=u_{t+1}+\gamma u_{t+2}+\dotsc+\gamma^{n-1}u_{t+n}+\gamma^n\nu_{t+n}$
